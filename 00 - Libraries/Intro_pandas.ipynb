### Ler dados do arquivo CSV

Para ler os dados do arquivo CSV isso será utilizado a função `pd.read_csv()`. O Pandas também consegue importar arquivos em diferentes extenções bem como Excel, SQL, Json mas para isto você deve mudar o final do comando  `pd.read_***` para o formato desejado (consultar no site oficial para entender melhor  como utilizar para outros formatos).  

No caso do exemplo abaixo, dentro da mesma pasta onde está o arquivo do código em *.ipynb* está também um arquivo com o nome *Dataset_CompanhiaMB.txt*. Esse arquivo não está com a extenção *.csv* mas está formatado como tal, basta abrí-lo no bloco de notas e notará que o conteúdo de cada coluna está separado por ";". Pelo padrão americano, se você usar somente o `pd.read_csv("Dataset_CompanhiaMB.txt")` ele vai entender que os dados estão sepados por `,` e que o separador decimal é o `.`.

Como nosso arquivo está separado por `;` é necessário informar isso usando `sep = ","`. Outro ponto que o pyton não entende a vírgula como separador decimal e neste caso também é necessário declarar por meio da expressão `decimal = ","` que o que separa as casas decimais no arquivo é `.` e não `,`. 

*Observação: o Python sempre interpreta o `.` como separador decimal. Ao declarar no pandas `decimal = ","` você só está indicando que o arquivo lido usa a `,` como separador decimal*



```python
tab2_1 = pd.read_csv("Dataset_CompanhiaMB.txt", sep =";", decimal = ",")
```
